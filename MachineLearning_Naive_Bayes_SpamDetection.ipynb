```python
#Function to display the output on the console and also write the output to a text file output.txt
def ProcessOutputToFile(data):
    print(data)
    f = open("output.txt", "a+")
    f.write(data + '\n')
    f.close()    
```

<b>TRAINING using the spambasetrain.csv dataset</b>

```python
#We will load all the data from the training set into a data frame and assign them labels
basedf = pd.read_csv('spambasetrain.csv', names=['char_freq_;', 'char_freq_(', 'char_freq_[', 'char_freq_!', 'char_freq_$', 'char_freq_#', 'capital_run_length_average', 'capital_run_length_longest', 'capital_run_length_total', 'ClassLabel'])

#We just display the first 4 records of the training data set
basedf.head()
```

```python
#Filter out the training dataset and just store all the data from the training set with Class Label 1 into a new training set
#known as dfClass1
dfclass1 = basedf[(basedf['ClassLabel'] == 1)]
dfclass1.head()
```

```python
dfclass0 = basedf[(basedf['ClassLabel'] == 0)]
dfclass0.head()
```

```python
dfclass1_Calc = dfclass1.describe()
dfclass1_Calc
```

```python
dfclass0_Calc = dfclass0.describe()
dfclass0_Calc
```

<b>Calculate the Prior for Both Class 0 and Class 1</b>

```python
priorOfClass0 = (dfclass0['ClassLabel'].count())/(basedf['ClassLabel'].count())
ProcessOutputToFile("Prior for Class 0 [P(C=0)] = " + str(priorOfClass0))
```

```python
priorOfClass1 = (dfclass1['ClassLabel'].count())/(basedf['ClassLabel'].count())
ProcessOutputToFile("Prior for Class 1 [P(C=1)]  = " + str(priorOfClass1))
```

<b>Calculate the Mean and Variance for both class 0 and class 1</b>

```python
mean_for_class0 = {}
mean_for_class1 = {}

variance_for_class0 = {}
variance_for_class1 = {}

dfCols =['char_freq_;', 'char_freq_(', 'char_freq_[', 'char_freq_!', 'char_freq_$', 
        'char_freq_#', 'capital_run_length_average', 'capital_run_length_longest', 'capital_run_length_total']

#function to calculate the mean for all the attributes of a given class
def calculateMean(meandict, classDf):
    for mail_attribute in dfCols:
        currSum = 0
        
        for index, row in classDf.iterrows():
            currSum += row[mail_attribute]
        
        meandict.update({mail_attribute : (currSum)/classDf[mail_attribute].count()})
        
#function to calculate the variance for all the attributes for a given class
def calculateVariance(varianceDict, meanDict, classDf):
    for mail_attribute in dfCols:
        currVar = 0
        
        for index, row in classDf.iterrows():
            currVar += np.power((row[mail_attribute] - meanDict[mail_attribute]), 2)
        
        varVal = currVar/(classDf[mail_attribute].count() -1 )
        varianceDict.update({mail_attribute : varVal})
        
    
#calculate mean for class 0
calculateMean(mean_for_class0, dfclass0)

#calculate mean for class 1
calculateMean(mean_for_class1, dfclass1)

#calculate variance for class 0
calculateVariance(variance_for_class0, mean_for_class0, dfclass0)

#calculate variance for classs 1
calculateVariance(variance_for_class1, mean_for_class1, dfclass1)

ProcessOutputToFile("Mean for class 0 " + str(mean_for_class0))
ProcessOutputToFile("variance for class 0 " + str(variance_for_class0))
ProcessOutputToFile("Mean for class 1 " + str(mean_for_class1))
ProcessOutputToFile("variance for class 1 " + str(variance_for_class1))
```

<b>TESTING using the spambasetest.csv</b>

```python
#Load the test data from spambasetest.csv into a dataframe
testDf = pd.read_csv('spambasetest.csv', names=['char_freq_;', 'char_freq_(', 'char_freq_[', 'char_freq_!', 'char_freq_$', 'char_freq_#', 'capital_run_length_average', 'capital_run_length_longest', 'capital_run_length_total', 'ClassLabel'])
testDf.head()
```

```python
#Function to calculate the gaussian distribution for a given attribute with a given mean and standarad deviation


def calcGaussianDist(x, mean, variance):
    val = (1/np.sqrt(2 * np.pi * variance)) * (np.exp((-1 * np.power(x - mean, 2)) / (2 * variance)))
    return val 

#Function to calculate the sum of all the PDFs of all the attributes for a given class
def calcEstPdf(row, meanList, varList):
    val=0
    for attrName in dfCols:
        attrGauss = calcGaussianDist(row[attrName], meanList[attrName], varList[attrName])
        val = val + np.log(attrGauss)
        
    return val

#List to store the results of all the predictions made
predictedResultSet=[]

#Counter to maintain the count of all the correct results predicted
correct=0

#Counter to maintain the count of all the results predicted incorrectly 
error=0

#Iterate thru the list of all the test cases to predict labels using naive bayes gaussian
for index, row in testDf.iterrows():
    #Calculate the estimated pdf for the given example for Class 0
    estPDFClass0 = calcEstPdf(row, mean_for_class0, variance_for_class0) + np.log(priorOfClass0)
    
    #Calculate the estimated pdf for the given example for Class 1
    estPDFClass1 = calcEstPdf(row, mean_for_class1, variance_for_class1) + np.log(priorOfClass1)
    
    #Calculate the class of the given example based on the estimated pdfs
    if (estPDFClass0 > estPDFClass1):
        classL = 0 
    else: 
        classL = 1
    
    #print("class 0 : " + str(estPDFClass0) + " class 1: " + str(estPDFClass1) + " Pred : "  + str(classL))
    
    #Check if the expected label for the given class and the  calculated label for the given class are 
    #the same, if not then we will count that as an error.
    if (row['ClassLabel'] == classL):
        correct = correct+1;
    else:
        error = error+1;
    
    #Store the predicted result for the given example in the result
    predictedResultSet.append(classL)
```

```python
ProcessOutputToFile("Predicted class for all test Examples : " + str(predictedResultSet))
```

```python
ProcessOutputToFile("Total Number of test examples classified correctly :  " + str(correct))
ProcessOutputToFile("Total Number of test examples classified incorrectly :  " + str(error))
```

```python
#Calculate the percentage of error on test examples
errorPercentage = ((error) / (error+correct))*100

#Calculate the percentage of results predicted correctly
correctPercentage = ((correct)/ (error+correct)) * 100
```

```python
ProcessOutputToFile("Percentage of error on the test examples : " + str(errorPercentage) + "%")
ProcessOutputToFile("Percentage of results predicted correctly on test examples : " + str(correctPercentage) + "%")
```

<h3> Calculation for finding accuracy using Zero-R </h3>

```python
'''
Incase if we are using the Zero-R method for predictions, based on our prior calculations
we find that the prior for P(C) for c=0 i.e 0.5981993998 is greater than the P(C) for c=1 
i.e 0.4018006002. Thus, based on Zero-R our classifier Class 0 for all the given test examples.
Therefore it will predict class for all the examples as class 0
'''

#However, we find that the based on the test dataset provided, the actual predictions for class zero were supposed to be
#just 118.
c0 = testDf[(testDf['ClassLabel'] == 0)]

#it will predicted all the samples as class 0 using zero-R , thus the total examples predicted as zero as
resultsPredictedCount = testDf['ClassLabel'].count() 

#Actual count of the predicted results as class zero is supposed to be as below
resultsActuallyPredictedForGivenClass = c0['ClassLabel'].count()

#Therefore, error in predcited results is as below
errorPredict = resultsPredictedCount - resultsActuallyPredictedForGivenClass
correctPredict =  resultsPredictedCount - errorPredict

print("Error in  prediction using zero-R is : "  + str((errorPredict/(errorPredict+correctPredict)) * 100) + "%")

#Thus, the accuracy of the given zero-r can be calculated using the below formula
accuracy = (correctPredict/ (errorPredict+correctPredict)) * 100
print("Accuracy of Zero-R method is : " + str(accuracy) + "%")
```
